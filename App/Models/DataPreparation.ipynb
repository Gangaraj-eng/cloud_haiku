Calculating Statistic for each team like average score of team, average wickets taken by team, average catches from the data of all the players in the team

```python
teams_data = pd.read_csv('../Dataset/ICC_23_Teams_List.csv')
players_data = pd.read_csv('../Dataset/ICC_Player_Stats.csv')
teams_data.shape,players_data.shape
```

```python
teams_data.head(10)
```

These are the list of teams that are participating in ICC world cup 2023 with their ODI Rankings and the number of world cup matches played.

```python
players_data.head()
```

These are the statistics of all the player in each team based upon their performance in number of ODI and World cup matches

```python
players_data.isnull().sum()
```

```python
# filling null values with average
nullData = players_data.isnull().sum()
for col in players_data.columns:
    if nullData[col]!=0:
        players_data[col].fillna(players_data[col].mean(), inplace=True)
```

```python
# grouping players by team
grouped_data = players_data.groupby('Team Name')
```

```python
avgData = grouped_data.mean()
# average score per match
teams_data['Avg Score']= avgData['Runs'].values/avgData['Matches'].values
teams_data['Avg Wickets'] = avgData['Wickets Taken'].values
teams_data['Avg Catches'] = avgData['Catches'].values
teams_data['Avg Fours'] = avgData['Fours'].values
teams_data['Avg Sixes'] = avgData['Sixes'].values
teams_data['Avg Half Centuries'] = avgData['Half Centuries'].values
teams_data['Avg Centuries'] = avgData['Centuries'].values
teams_data.head(10)
```

```python
teams_data.to_csv('../Dataset/ICC_Team_Stats.csv',index=False)
```

Scrapping ICC matches data from 1999-2019

```python
# scrapping matches data of icc world cups
import requests
from bs4 import BeautifulSoup
from datetime import datetime

def get_date(div):
    try:
        date = div.find_all('div','schedule-date')[0]
        ind2 = str(date).find('|')
        ind1 = str(date).rfind('"',0,ind2)
        ts = datetime.fromtimestamp(int(str(date)[ind1+2:ind2])/1000)
        date = datetime.strftime(ts, '%Y/%m/%d')
        return date
    except:
        return None

def getTeams(div):
    teams = div.find_all('div','cb-col-60 cb-col cb-srs-mtchs-tm')[0]
    a1 = teams.find_all('a','text-hvr-underline')
    a2 = a1[0].text.split(" vs ")
    team1 = a2[0]
    team2 = a2[1].split(',')[0]
    team1 = ' '.join([x.capitalize() for x in team1.split(' ')])
    team2 = ' '.join([x.capitalize() for x in team2.split(' ')])
    wt = teams.find_all('a','cb-text-complete')[0].text.split(' won ')[0]
    venue = teams.find_all('div','text-gray')[0].text
    return team1,team2, wt, venue

matches = pd.DataFrame()
cnt=0
def extract_data(url,season):
    # Send a GET request to the URL
    response = requests.get(url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the HTML content of the page
        soup = BeautifulSoup(response.text, 'html.parser')
        
        previous_date = None
        matches_div = soup.find_all('div','cb-series-matches')
        
        for match in matches_div:
            global cnt
            date = get_date(match)
            team1, team2, winningTeam, venue = getTeams(match)
            if date is None:
                date = previous_date
            matches.loc[cnt,['season','team1','team2','date','winner','Venue']]=[season,team1,team2,date,winningTeam,venue]
            cnt = cnt+1
            previous_date = date

    else:
        print(f"Failed to retrieve the page. Status Code: {response.status_code}")

extract_data("https://www.cricbuzz.com/cricket-series/2697/icc-cricket-world-cup-2019/matches",'2019')
extract_data("https://www.cricbuzz.com/cricket-series/2223/icc-cricket-world-cup-2015/matches",'2015')
extract_data("https://www.cricbuzz.com/cricket-series/228/icc-world-cup-2011/matches",'2011')
extract_data("https://www.cricbuzz.com/cricket-series/905/icc-world-cup-2007/matches",'2007')
extract_data("https://www.cricbuzz.com/cricket-series/796/icc-world-cup-2003/matches",'2003')
extract_data("https://www.cricbuzz.com/cricket-series/718/icc-world-cup-1999/matches",'1999')
matches.head()
```

```python
d1 = pd.read_csv('../Dataset/ICC_matches_stats.csv')
combined = pd.concat([d1, matches], axis=0)
combined.shape
combined.to_csv('../Dataset/ICC_matches_stats.csv',index=False)
```
